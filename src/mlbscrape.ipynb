```python
import numpy as np
import pandas as pd
```

```python
seasons = [str(season) for season in range(2012,2024)]
teams = ['ARI','ATL','BAL','BOS','CHC','CHW','CIN','CLE','COL','DET',
         'HOU','KCR','LAA','LAD','MIA','MIL','MIN','NYM','NYY','OAK','PHI',
         'PIT','SDP','SEA','SFG','STL','TBR','TEX','TOR','WSN']
print(f'number of teams={len(teams)}')
```

```python
import random
import time

mlb_df = pd.DataFrame()

for season in seasons:
    for team in teams:
        url = 'https://www.baseball-reference.com/teams/' + team + '/' + season + '-schedule-scores.shtml'
        print(url)

        schedule_df = pd.read_html(url, header=1, attrs={'id': 'team_schedule'})[0]

        # Insert metadata columns
        schedule_df.insert(0, "Season", season)
        schedule_df.insert(1, "Team", team.upper())

        # Append to main DataFrame
        mlb_df = pd.concat([mlb_df, schedule_df], ignore_index=True)

        time.sleep(random.randint(7, 8))
```

```python
mlb_df.to_csv('mlb_scraper.csv', index=False)
```
